### Installation Phase 3: Install and Import Matplotlib

To create a graph, create a new Python file. To do this: 
1. Open PyCharm Community Edition, and select File. 
2. From **File**, select **New Project**. 
3. To use libraries in Python files, you must first install them. In the newly created Python project, navigate to the bottom of the application window, and select Python Packages.
4. In the search window that follows, type Matplotlib. Select the library, and then select Install. 

PyCharm will now install the Matplotlib library. To activate and use Matplotlib within a Python file, import Matplotlib in the Python file/code. To accomplish this, type in:


```python
import matplotlib.pyplot as plt 
```

This will import the Matplotlib library for use within Python. You can access all functions in this library with **plt**.


### Create Data

To create data for use with Matplotlib, enter the data as a list. A **list** is a data type that stores a list of values. Create two lists (for both the X and Y axis of the graph) and enter values separated by commas. 

To do this, copy the line below, and alter the values within the list as needed:

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]
```

The letters “x” and “y” denote the names of the list, and the contents of each list are within the brackets, between the commas.

### Create the XY Graph

Matplotlib contains a variety of functions that create and interpret graphs. To create a graph, use the **plot** function within the Matplotlib library, and type in:


```python
plt.plot(x,y)
```

“plt.plot” accesses the plot function in the Matplotlib library, and x and y both denote the names of the two lists, the x and y-axis of the graph.


### Set Graph Limits

A graph limit ensures that no extremely large values distort the display of the graph. To set limits on the x and y values of the graph, use the “**xlim**” and “**ylim**” functions. To use these functions, type the following:


```python
xmax = 6
ymax = 55
plt.xlim(1, xmax)
plt.ylim(10, ymax)
plt.plot(x,y)
```

Note: replace the values of “**xmax**” and “**ymax**” with any integer of choice that denotes the maximum desired values of the x and y-axis. Repeat the same process and replace "0" with a desired minimum.


### Label the Graph

To label the x and y axis of the graph, use the “**xlabel**” and “**ylabel**” functions of the graph. To create a title for the graph, use the “**title**” function. To utilize these functions, type in the following:


```python
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Simple Graph')
plt.plot(x,y)
```

“X-axis Label,” “Y-axis Label,” and “Simple Graph” correspond to the labels of the x-axis, y-axis, and the title of the graph. To change the names of these labels/titles, replace the placeholder titles with the desired names.


### Display the Graph

All of the previously written code in this guide creates a graph with proper labels and with two sets of data. However, in each step, the **plot** function displays the graph. To display the graph at any point in time, type in:

```python
plt.plot(x,y)
```

This will display the graph, with x as the x-axis, and y as the y-axis.

### Save the Graph

To save the graph for use outside of the Python project, save the graph locally onto the device. Save the graph by running the “**savefig**” function, and type in:


```python
plt.plot(x,y)
plt.savefig('simple_graph.png')
```

Alter the name of the file as desired; to do this change “**simple_graph.png**” to the desired name of the file. This completes the creation of a graph in Python using the Matplotlib library.
